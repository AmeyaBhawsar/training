```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Averaging Kernel

```python
img1 = cv2.imread("automobile.png")

kernel = np.ones((5,5), np.float32) / 25

sm_img = cv2.filter2D(img1, -1, kernel)

cv2.imshow('Original Image', img1)
cv2.imshow('Smoothed Image', sm_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_image = np.hstack([img1, sm_img])
cv2.imshow('Both Images', new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 2. Gaussian Kernel

```python
kernel_size = 7 # Adjust as needed
sigma = 0.35 * ((kernel_size - 1) * 0.5 - 1) + 0.9 # Gaussian Standard Deviation
kernel = cv2.getGaussianKernel(kernel_size, sigma)

print(sigma, kernel)

smoothed_image = cv2.filter2D(img1, -1, kernel)

new_image = np.hstack([img1, smoothed_image])
cv2.imshow('Both Images', new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 3. Median Kernel

```python
gray = cv2.imread('automobile.png', 0)

print(gray.shape)

kernel_size = 9  # More the size, more the blur
median = cv2.medianBlur(gray, kernel_size)

plt.figure(figsize = (10, 6))
plt.subplot(121), plt.imshow(gray, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(median, cmap='gray'), plt.title('Median Filtered Image')
```

```python
gray2 = cv2.imread('automobile.png')
kernel_size = 15
median = cv2.medianBlur(gray2, kernel_size)

plt.figure(figsize = (10, 6))
plt.subplot(121), plt.imshow(gray2, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(median, cmap='gray'), plt.title('Median Filtered Image')
```
